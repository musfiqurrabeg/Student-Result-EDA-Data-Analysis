```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

## Drop the unnamed column ##

```python
df = df.drop("Unnamed: 0", axis=1)

df.head()
```

## Gender Distribution ##

```python
plt.figure(figsize=(8, 5))
ax = sns.countplot(x="Gender", data = df, palette="Set3")
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```

```python
# From the above chart we have analysed that
# The number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore" : 'mean', "ReadingScore" : 'mean', "WritingScore" : 'mean'})

gb
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(data= gb, annot=True)
plt.title("Relationship Between Parent's Education And Student's Education")
plt.show()
```

### From the chart, we have concluded that the education of the parents have a good impact ###

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore" : 'mean', "ReadingScore" : 'mean', "WritingScore" : 'mean'})

gb1
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(data= gb1, annot=True)
plt.title("Impact of Parent's Marital Status on Student's Score")
plt.show()
```

### From the chart, we have concluded that, there is no negative impact on student's score due to their parent's marital status ###

### Let's check how much sports impact on student score

```python
plt.figure(figsize=(8, 6))
ax = sns.countplot(data=df, x="PracticeSport", palette="Set3")
for i in ax.containers:
    ax.bar_label(i)
plt.title("Practice Sport")
plt.show()
```

```python
gb2 = df.groupby("PracticeSport").agg({"MathScore" : 'mean', "ReadingScore" : 'mean', "WritingScore" : 'mean'})

gb2
```

## Distribution of Ethnic Group ##

```python
unique_groups = df['EthnicGroup'].unique()

unique_groups
```

```python
groupA = df.loc[df['EthnicGroup'] == 'group A'].count()
groupB = df.loc[df['EthnicGroup'] == 'group B'].count()
groupC = df.loc[df['EthnicGroup'] == 'group C'].count()
groupD = df.loc[df['EthnicGroup'] == 'group D'].count()
groupE = df.loc[df['EthnicGroup'] == 'group E'].count()

plt.figure(figsize=(7,7))
groupname = ["group A", "group B", "group C", "group D", "group E"]
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#D4A6FF']
grouplist = [groupA['EthnicGroup'], groupB['EthnicGroup'], groupC['EthnicGroup'], groupD['EthnicGroup'], groupE['EthnicGroup']]
plt.pie(grouplist, labels = groupname, autopct='%.2f%%', shadow=True, colors=colors)
plt.title("Distribution of Ethnic Groups")
plt.show()
```
